```python
import numpy as np

np.percentile(df_part['pressure'], [25, 50, 75])

# 25th: 25% of the data is below this value. Lower than this is considered low pressure.
# 50th: 50% of the data is below this value, this is the median value.
# 75th: 75% of the data is below this value. Higher than this is considered high pressure.
```

Calculate standard deviation of the data. Show this number.

```python
stdev = np.std(df_part['pressure'])
print(stdev)
```

Multiple the standard deviation by 2. Count the number of values in the pressure column that are:

* Smaller than the **median - 2 * stddev**
* Bigger than the **median + 2 * stddev**

You've just looked at outliers. They are not part of the course material for this year, but playing around with pandas dataframes (calculating values, filter columns, counting number of values) is.

```python
median = np.median(df_part['pressure'])

print("Nr below: ", len(df_part[df_part['pressure'] < (median-2*stdev)]))
print("Nr above: ", len(df_part[df_part['pressure'] > (median+2*stdev)]))
```
